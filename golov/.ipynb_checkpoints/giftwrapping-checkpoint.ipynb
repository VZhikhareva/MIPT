```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gen_points(n):
    X = np.random.normal(0, 1, n)
    Y = np.random.normal(0, 1, n)
    return X, Y
```

```python
pp = gen_points(10000)
points = {'X': pp[0], 'Y': pp[1]}
```

```python
def first_point_index(points = points):
    '''
    looking for a first vertex
    '''
    min_x_arg = points['X'].argmin()
    min_y_arg = np.min(np.where(points['Y'] == points['Y'][min_x_arg]))
    if min_x_arg == min_y_arg:
        return min_x_arg
    else:
        return None
```

```python
## checking
plt.scatter(points['X'], points['Y'], s = 1)
plt.scatter(points['X'][first_point_index()], points['Y'][first_point_index()], c = 'red')
```

```python
p1_ind = first_point_index() # here is also minimal x
SHIFT_X = points['X'][p1_ind]
## moving to a positive-X domain 
points['X'] = (points['X'][:] - SHIFT_X)/8
# points['Y'] = (points['Y'][:] - y_min)/8

plt.scatter(points['X'], points['Y'], s = 1)
plt.scatter(points['X'][p1_ind], points['Y'][p1_ind], c = 'red')
```

```python
def second_point(points = points, p1_ind = first_point_index()):
    '''
    looking for a second vertex
    '''
    max_angle = 0
    p2_ind = -1
    for i in range(len(points['X'])):
        tmp = np.angle(points['X'][i] + 1.0j*(points['Y'][i] - points['X'][p1_ind]) - 1.0j*(points['Y'][p1_ind]))
        if (i != p1_ind) and (tmp > max_angle) and (points['Y'][i] >= points['Y'][p1_ind]):
            max_angle = tmp 
            p2_ind = i
    return p2_ind, max_angle
```

```python
## checking
plt.scatter(points['X'], points['Y'], s = 1)
plt.scatter(points['X'][first_point_index()], points['Y'][first_point_index()], c = 'red', s = 20)
plt.scatter(points['X'][second_point()[0]], points['Y'][second_point()[0]], c = 'green', s = 20)
# plt.scatter(1, 0, c = 'blue')
```

```python
def cos_btw_fragments(center_x, center_y, previous_x, previous_y, next_x, next_y):
    '''
    calculation of cos of angle that is between two fragments, where center is their common vertex
    '''
    a = [previous_x - center_x, previous_y - center_y]
    b = [next_x - center_x, next_y - center_y]
    return np.dot(a, b)/np.linalg.norm(a)/np.linalg.norm(b)
```

```python
passed_points_ind = [first_point_index(), second_point()[0]]  # all points that were included in shell already
```

```python
indeces_upd = set(np.arange(len(points['X']))).difference(set(passed_points_ind))
```

```python
passed_points_ind = [first_point_index(), second_point()[0]]  # all points that were included in shell already

def next_point(points = points, passed_points_ind = passed_points_ind):
    indeces_upd = set(np.arange(len(points['X']))).difference(set(passed_points_ind[1:])) 
    tmp_ind = p1_ind
    cos_value = 1000
    for item in indeces_upd:
        tmp = cos_btw_fragments(center_x = points['X'][passed_points_ind[-1]], center_y = points['Y'][passed_points_ind[-1]]
                                , previous_x = points['X'][passed_points_ind[-2]], previous_y = points['Y'][passed_points_ind[-2]]
                                , next_x = points['X'][item], next_y = points['Y'][item])
        if tmp < cos_value:
            cos_value = tmp
            tmp_ind = item
    passed_points_ind.append(tmp_ind)
    return cos_value
```

```python
while passed_points_ind[-1] != passed_points_ind[0]:
    next_point()
```

```python
## moving back all the points along x axes

points['X'] = (points['X'][:]*8 + SHIFT_X)
```

```python
plt.scatter(points['X'], points['Y'], s = 1, color = 'c')
plt.plot(points['X'][passed_points_ind], points['Y'][passed_points_ind], c = 'green')
```
